```python
#set up train_gen and val_gen using ImageDataGenerator,
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

train_gen = datagen.flow_from_directory(
    'data/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    'data/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
# Build and Compile the CNN Model
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.optimizers import Adam

# Load the base MobileNetV2 model
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False

# Add custom classification head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
output_layer = Dense(train_gen.num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output_layer)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#Train the Model
history = model.fit(
    train_gen,
    epochs=10,
    validation_data=val_gen
)
```

```python
#Evaluate the Model
val_loss, val_acc = model.evaluate(val_gen)
print(f'Validation Accuracy: {val_acc * 100:.2f}%')
```

```python
#Save the model
model.save('illegal_fish_classifier.h5')
```

```python
from tensorflow.keras.preprocessing import image  #  add this if not already done
import numpy as np  # also make sure this is imported

def predict_image(img_path, threshold=0.75):
    # Load and preprocess image
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)

    # Predict
    prediction = model.predict(img_array)
    predicted_index = np.argmax(prediction)
    predicted_class = list(train_gen.class_indices.keys())[predicted_index]
    confidence = prediction[0][predicted_index]

    print(f"Predicted Fish: {predicted_class}")
    print(f"Confidence: {confidence:.2f}")

    if confidence >= threshold:
        print(" Status: ILLEGAL fish detected.")
    else:
        print(" Status: Likely LEGAL or UNKNOWN fish.")
```
